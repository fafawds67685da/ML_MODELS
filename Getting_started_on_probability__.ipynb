# visualization of different Outcomes in a dice

```python
import numpy as n
import matplotlib.pyplot as m

n.random.randint(1,7)
```

```python
few_rolls = n.random.randint(1,7,10)
many_rolls = n.random.randint(1,7,1000)

few_counts = n.histogram(few_rolls, bins=n.arange(.5,7.5))[0]
many_counts = n.histogram(many_rolls, bins=n.arange(.5,7.5))[0]

print(few_counts) # frequency of every face of dice where there were 10 outcomes
print(many_counts) # frequency of every face of dice where there were 100 outcomes
```

```python
fig, (ax1, ax2) = m.subplots(2, 1, figsize =(8,3)) # plotted graphy for the distrubutuin to show the data was distributed uniformly
ax1.bar(n.arange(1,7), few_counts)
ax2.bar(n.arange(1,7), many_counts)
```

```python
fig, (ax1, ax2) = m.subplots(1, 2, figsize =(8,3)) # plotted graphy for the distrubutuin to show the data was distributed uniformly
ax1.bar(n.arange(1,7), few_counts)
ax2.bar(n.arange(1,7), many_counts)
```

# Probability distribution

```python
import scipy.stats as ss

b= ss.distributions.binom
for flips in [5, 10, 20, 40, 80]:
  success = n.arange(flips)
  our_distribution = b.pmf(success, flips, .5)
  m.hist(success, flips,weights=our_distribution)
```
